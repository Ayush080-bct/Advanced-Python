### Normaliztion


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x=np.array([[10],[20],[30],[40],[100]])
scaler=StandardScaler()
X_transformed=scaler.fit_transform(x)
print(X_transformed)
```

```python
x=np.array([[10],[20],[30],[40],[100]])
scaler=MinMaxScaler(feature_range=(0,1))
X_transformed=scaler.fit_transform(x)
print(X_transformed)
```

```python
x=np.array([[10],[20],[30],[40],[100]])
j=len(str(np.max(np.abs(x))))
X_transformed=x/10**j
print(X_transformed)
```

### Feature Encoding Types
- nominal: no order  
- ordinal: defined order

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cities=['Kathmandu','Pokhara','Biratnagar']
encoded=le.fit_transform(cities)
print(encoded)
```

```python

from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[['poor','fair','good']])# 1row 3 column
quality=[['poor'],['fair'],['good']]#2d array (list of list) 3 row 1 column
encoded=encoder.fit_transform(quality)
print(encoded)
```

```python
import pandas as pd
df=pd.DataFrame({'source':['river','lake','well']})
df_encoded=pd.get_dummies(df,columns=['source'])
print(df_encoded)
```

```python
import numpy as np
import time
x=np.random.random(10000000)
print(x[:5])
start=time.time()
for i in range(x.shape[0]):
  x[i]=x[i]+2
end=time.time()
print(x[:5])
print(f'Time taken is {end-start} seconds')
```
