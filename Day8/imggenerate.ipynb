# Chat bot

```python
import requests
from pprint import pprint
import base64#Base64 encodes arbitrary binary data into a restricted set of ASCII characters,
# which are represented as bytes (binary) during transmission.
import os
from dotenv import load_dotenv
load_dotenv()
api_key=os.getenv("api")

url = "https://router.huggingface.co/hf-inference/models/stabilityai/stable-diffusion-xl-base-1.0"
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type":"application/json"
}
#if content -type was binary we use base64 to encode
i=0
while True:
    prompt = input("Write prompt (or 'exit' to exit the chatbot): ")

    if prompt.lower() == "exit":
        break

    data = {
        "inputs": prompt
    }

    response = requests.post(url, headers=headers, json=data)
    content_type = response.headers.get("Content-Type", "")

    print("Status:", response.status_code)
    print("Content-Type:", content_type)

    if response.status_code != 200:
        print(response.text)
        continue

    #Case1 Raw image
    if content_type.startswith("image/"):
        ext = content_type.split("/")[-1]
        filename = f"output{i}.{ext}"

        with open(filename, "wb") as f:
            f.write(response.content)

        print(f"Image saved as {filename}")

    #Case2 JSON+base64
    elif "application/json" in content_type:
        result = response.json()

        if isinstance(result, list):
            image_base64 = result[0]["image"]
        else:
            image_base64 = result["data"][0]["b64_json"]

        image_bytes = base64.b64decode(image_base64)

        with open("o.png", "wb") as f:
            f.write(image_bytes)

        print("Image saved as output.png")

    else:
        print("Unknown response format")
        print(response.text)
    i=i+1
```
