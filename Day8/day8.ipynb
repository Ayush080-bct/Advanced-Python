```python
import requests
import pandas as pd
import pprint 
```

```python
city="Kathmandu"
api_key='e6ea5fb4ce5dd3fe497b1bba0b1c8ec1'
url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
response=requests.get(url)
pprint.pprint(response.json())
```

```python
from pandas import json_normalize
df=json_normalize(response.json())
print(df.info())
```

```python
df_weather=df[['name','main.temp','main.humidity','wind.speed','weather']]
df_weather.head()
```

```python
df_weather['description']=df_weather['weather'].apply(lambda x:x[0]['description'])
df_weather=df_weather.drop(columns=['weather'])
df_weather.head()
```

```python
city_names=['Biratnagar','Janakpur','Biratnagar','Kathmandu','New York','California']
final_list=[]
for citys in city_names:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={citys}&appid={api_key}&units=metric"
    r=requests.get(url)
    df=json_normalize(r.json())
    df_weather=df[['name','main.temp','main.humidity','wind.speed','weather']]
    df_weather['description']=df_weather['weather'].apply(lambda x:x[0]['description'])
    df_weather=df_weather.drop(columns=['weather'])
    final_list.append(df_weather)
pprint.pprint(final_list)

    
    
```

```python
final_df=pd.concat(final_list,ignore_index=True)
final_df.head()
final_df.to_csv('weather.csv',index=False)
```
