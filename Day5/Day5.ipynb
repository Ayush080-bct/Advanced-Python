```python
dataframe=pd.DataFrame([
{"Roll":1,"Name":"Ram","Eng":80,"Math":81,"science":82},
{"Roll":2,"Name":"shyam","Eng":79,"Math":80,"science":83},
])
print(dataframe)
```

```python
dataframe['Total']=dataframe['Eng']+dataframe['Math']+dataframe["science"]
print(dataframe)
```

```python
dataframe['Percentage']=(dataframe['Total']/3).round(decimals=3)
print(dataframe)
```
