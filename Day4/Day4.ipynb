```python
#Base class
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print(f"{self.name} is speaking")
#Derived class
class Dog(Animal):
    def __init__(self, name,breed):
        super().__init__(name)#super ->function which always point to upper 
        self.breed=breed
    def speak(self):#function overriding -> dyanmic i.e run time polymorphism but function overloading is compile time polymorphism
        super().speak()
        print(f"{self.name} of {self.breed} is Barking!")
d1=Dog("Lucy","Shiba Inu")
d1.speak()
```

```python
#Multiple inheritance
class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
class c(A,B):#c class inheritate A first so method in class A is called
    pass
c1=c()
print(c1)
c1.show()
```

```python
#Method resolution order
c.mro()
```

```python
class DataList:
    def __init__(self,aList):
        self.aList=aList
    def Mean(self):
        return sum(self.aList)/len(self.aList)
l1=DataList([1,2,3,4,5,6])
l1.Mean()        
```

```python
while(True):
    
    try:
        n1=int(input("Enter the First Number: "))
        n2=int(input("Enter the second number: "))
        result=n1/n2
        print("Result : ",result)

    except ZeroDivisionError:
        print("cannot Divide by zero!")
    except ValueError:
        print("Please enter the valid number")
    except KeyboardInterrupt:
        break
    finally:
        print("This always run")
```

# Logging

while debugging using pdb we pause the program ,instead of stopping execution ,you should start recording program behaviour.
Used heavily in backend system and production code 

```python
import logging
logging.basicConfig(level=logging.DEBUG)
logging.debug("Starting Program")
x=10
y=0
logging.debug(f"x={x} , y={y}")
try:
    print(x/y)
except Exception as e:
    logging.error(e)
```
