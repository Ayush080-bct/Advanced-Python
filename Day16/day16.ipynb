```python
import schedule
import time
from datetime import datetime

def write_files():
    with open("schedule.txt", "a") as f:
        f.write(f"Hello! I am Acer assistant and The time is {datetime.now()}\n")
schedule.every(1).seconds.do(write_files)
while True:
    schedule.run_pending()
    time.sleep(1)
```
