```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,30,0.27)
y=1.223*np.square(x)+2.456
plt.figure(figsize=(10,8))
plt.plot(x,y,c='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.title('x vs y')
plt.grid()
plt.show()
```

```python
# PLOTs using Subplots
fig,ax=plt.subplots()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,y,c='r')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("x vs Y")
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x,y,c='g')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("x vs Y")
plt.tight_layout
```

```python

fruits=['Apple','Guava','Banana','Orange']
quantity=[120,90,70,150]
fig,axes=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i,j].bar(fruits,quantity,color=["Red",'Gray',"Limegreen","Magenta"],alpha=0.77)
        axes[i,j].set_title(f'Plot {i},{j}')
        axes[i,j].set_xticks(ticks=fruits)
        axes[i,j].set_xticklabels(labels=fruits,rotation=45)
plt.tight_layout()

# plt.subplots_adjust(wspace=0.25,hspace=0.25)
```

```python
np.random.seed(42)

data=np.random.normal(loc=50,scale=10,size=1000)
plt.hist(data,bins=20,color='skyblue',edgecolor='black')
plt.xlabel("value")
plt.ylabel('frequency')
plt.title("Histogram of Random data")
plt.show()
```

```python
from sklearn.datasets import make_moons
X,y=make_moons(
    n_samples=1000,
    noise=0.04,
    random_state=42
)
print(X.shape)
fig,axes=plt.subplots()
axes.scatter(X[:,0],X[:,1],c=y)
axes.text(x=1,y=1,s="Aakash jogbani",family="monospace",fontsize=10)
axes.set_title('Scatter plot')
plt.savefig("Scatter.png")
```

```python
import pandas as pd

from sklearn.datasets import load_iris
data=load_iris()
X,y=load_iris(as_frame=True,return_X_y=True)

feture_names=data.feature_names
class_names=data.target
```

```python
fig,axes=plt.subplots(2,2)
axes[0,1].scatter(
    X['sepal length (cm)'], 
    X['petal length (cm)'], 
    c=y, 
    cmap='viridis'
)
axes[0,1].set_xlabel("Sepal length (cm)")
axes[0,1].set_ylabel("Petal length (cm)")
axes[0,1].set_title("Sepal vs Petal Length")

axes[0,1].set_title("Bar Graph",fontsize=8)
axes[0,0].bar(x=['setosa','Versicolor','Virginica'],
              height=pd.Series(y).value_counts(),color=['Red','Green','Blue'])
axes[0,0].set_title("Bar graph")
axes[1,0].hist(X['sepal width (cm)'],bins=20,color='c',edgecolor='black',alpha=0.7)
axes[1,0].set_title('Histogram of Sepal Width',fontsize=8)
axes[1,1].boxplot(X['petal width (cm)'],vert=False)
axes[1,1].set_title('Box Plot of Petal Width',fontsize=8)
plt.tight_layout()
```
