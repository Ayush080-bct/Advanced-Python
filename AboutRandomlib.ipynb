# Random
*The primary Python library for generating random numbers and making random selections is the built-in random module.*
*It is part of the Python Standard Library, meaning no additional installation is required.*

```python
import random#	Returns a random floating-point number between 0.0 (inclusive) and 1.0 (exclusive).
#inclusive means that the number is possible outcome ,and exclusive means 1 is not possible outcome <1 and >0is possible outcome 
print(random.random())#generate floating value
```

```python
random.randint(1,10)#Returns a random integer N such that a(1) <= N <= b(4) (both inclusive).#return random integer
```

```python
#gives unique number
myset=set()
for _ in range(10):

    myset.add(random.randint(10,20))#both possible outcome
print(myset)
```

```python
mylist=[]
for i in range(10):
    
    number=random.randint(10,20)#both possible outcome as inclusive
    mylist.append(number)
print(mylist)
```

```python
random.uniform(10,20)#return float value and a is inclusive and b is exclusive i.e 20 is not possible outcome
```

```python
random.uniform(1.5,1.51)
```

```python
random.choice(mylist)#Returns a random element from a non-empty sequence (list, tuple, string, etc.
```

```python
random.choice(['Ayush',"Ram","shyam"])
```

```python
random.shuffle(mylist)#	Shuffles the elements of a list 'mylist' in place (modifies the original list).
```

```python
random.sample(range(1,1000),5)#Returns a list of k(which is 5 for now)
#unique random elements chosen from the population(range(1,1000)) without replacement.
```

```python
random.sample(range(1,1000),10)
```
