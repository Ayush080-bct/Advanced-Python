```python
from sqlalchemy import create_engine

engine = create_engine(
    "sqlite:///C:/Users/Acer/Desktop/Advanced Python/Day6/students.db",
    echo=True
)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String

Base = declarative_base()

class Student(Base):
    __tablename__ = 'students'

    id = Column(Integer, primary_key=True)
    name = Column(String(50))
    age = Column(Integer)
    address = Column(String(100))

    def __repr__(self):
        return f"Student(id={self.id}, name={self.name}, age={self.age}, address={self.address})"
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker 
Session = sessionmaker(bind=engine)
session = Session()
```

```python
student1 = Student(name='Ram', age=20, address='Ithari')
student2 = Student(name='Shyam', age=19, address='Ithari')
student3 = Student(name='Hari', age=20, address='Biratnagar')

session.add_all([student1, student2, student3])
session.commit()
```

```python
data = session.query(Student).all()
for row in data:
    print(row)
```
