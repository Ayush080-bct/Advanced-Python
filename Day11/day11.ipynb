### Normaliztion


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x=np.array([[10],[20],[30],[40],[100]])
scaler=StandardScaler()
X_transformed=scaler.fit_transform(x)
print(X_transformed)
```

```python
x=np.array([[10],[20],[30],[40],[100]])
scaler=MinMaxScaler(feature_range=(0,1))
X_transformed=scaler.fit_transform(x)
print(X_transformed)
```

```python
x=np.array([[10],[20],[30],[40],[100]])
j=len(str(np.max(np.abs(x))))
X_transformed=x/10**j
print(X_transformed)
```

### feature encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cities=['Kathmandu','Pokhara','Biratnagar']
encoded=le.fit_transform(cities)
print(encoded)
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[['poor','fair','good']])# 1row 3 column
quality=[['poor'],['fair'],['good']]#2d array (list of list) 3 row 1 column
encoded=encoder.fit_transform(quality)
print(encoded)
```
